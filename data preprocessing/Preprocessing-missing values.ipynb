
# Descriptive Statistics

#Descriptive Statistics mein, hum data ki descriptive statistics ko calculate kar ke missing values ko identify karte hain.



```python
# calculate missing values
print("----------------------------------------")

data.isnull().sum().sort_values(ascending = False)
```

```python
print("Percentage of every column")
round(data.isnull().sum() / len(data) * 100, 2).sort_values(ascending = False)
```

```python
#Mean, Median, Ya Mode Se Data Ko Impute Karna

#mean
```

```python
#---mean missing value handling

# missing value 
NAN = data.isnull().sum().sort_values(ascending = False)

print(f"Missing values:\n", NAN)

# Calculate the mean
mean = data['age'].mean()
print(f'percentage of missing values in age',mean)
# Replace the missing values with mean

data['age'] = data['age'].fillna(mean)
print("----------------------------------------")
print("-----------Without Missing Values---------\n")
print(f"Without missing values: \n", data)
```

```python
data.head()
```

```python
#. Median

median = data['age'].median()
```

```python
median
```

```python
data['age'] = data['age'].fillna(median)
```

```python
print("Filling Nan's with median in age missing values:\n ", data.head())
```

```python
#------Mode

#calculating mode
mode = data['age'].mode()

# replacing with nan values

data['age'] = data['age'].fillna(mode)


# print values 

data.head()
```

```python
# import Libarires

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import KNNImputer

#load data
data = sns.load_dataset('titanic')

# missing values
data.isnull().sum().sort_values(ascending = False)

# Initialize the KNNImputer
imputer = KNNImputer(n_neighbors = 2)

# Replace the missing values with KNN imputation
data['age'] = imputer.fit_transform(data[['age']])

print("----------------------------------------")
# Print the data without missing value

data.head(4)
```

# Deep Learning Techniques Ka Istemal
Deep learning techniques jaise autoencoders bhi missing values ko handle karne mein madadgar sabit ho sakte hain.

```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

# load data

data = sns.load_dataset('titanic')

data.isnull().sum().sort_values(ascending = False)


# Initialize the IterativeImputer

imputer = IterativeImputer()

# Replace the missing values with deep learning techniques
data['age'] = imputer.fit_transform(data[['age']])

print("-Print the data without missing value-")

data.head()
```

# Simply Delete Kar Dena

```python
# load data

data = sns.load_dataset('titanic')

data.isnull().sum().sort_values(ascending = False)


# Delete the rows with missing values
data = data.dropna()

# print after droping missing values

data.head()
```
